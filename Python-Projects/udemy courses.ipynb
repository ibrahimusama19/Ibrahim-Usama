#  Which Course Has More Lectures (Free or Paid)? 

```python
data.columns
```

```python
sns.barplot(x="is_paid",y="num_lectures",data=data)
plt.xlabel("is paid ?")
plt.ylabel("number of lectures")
plt.show
```

 # Which Courses Have A Higher Number of Subscribers Free or Paid?

# Which Level Has The Highest Number of Subscribers?

# Display 5 Most Popular Courses As Per Number of Subscribers

# Does Price Affect the Number of Reviews?

# Display 10 Most Popular Python Courses As Per Number of Subscribers

```python
top_10_pythonCourses=data[data["course_title"].str.contains("python",case=False)].sort_values(by="num_subscribers",ascending=False).head(10)
```
